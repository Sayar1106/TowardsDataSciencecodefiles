## Styling

```python
fifa_df[['Value', 'Wage']] = fifa_df[['Value', 'Wage']].apply(lambda s: 
    s.replace('[\€,)]','', regex=True))
```

```python
fifa_df['Value'] = fifa_df['Value'].replace({'K':'*1e3', 'M':'*1e6'}, regex=True).map(pd.eval).astype(float)
fifa_df['Wage'] = fifa_df['Wage'].replace({'K':'*1e3', 'M':'*1e6'}, regex=True).map(pd.eval).astype(float)
```

```python
fifa_df.head(100).sample(100).style.format({"Value": "€{:20,.0f}", "Wage": "€{:20,.0f}"})\
                 .bar(subset='Value', color='lightblue')\
                 .bar(subset='Wage', color='orange')\
                 .background_gradient(subset='Overall', cmap='winter_r')\
                 .applymap(lambda x: f"color: {'red' if x >= 30 else 'black'}", subset='Age')
```
